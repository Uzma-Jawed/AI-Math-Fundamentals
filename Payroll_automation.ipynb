**ðŸŽ¯ Goal: Automate Payroll Processing**

* We want a Python script that:

* Takes employee data (name, hours worked, hourly rate)

* Calculates total pay (with optional tax deduction)

* Generates a basic payroll summary


---



**Employee Data**

Start with a list of dictionaries â€” each dictionary is one employee.

```python
employees = [
    {"name": "Uzma", "hours_worked": 40, "hourly_rate": 500},
    {"name": "Ali", "hours_worked": 35, "hourly_rate": 600},
    {"name": "Sara", "hours_worked": 45, "hourly_rate": 550}
]
```

**Define a Function to Calculate Pay**

This function calculates gross pay and tax deduction.

```python
def calculate_pay(hours, rate, tax_rate=0.10):
    gross_pay = hours * rate
    tax = gross_pay * tax_rate
    net_pay = gross_pay - tax
    return gross_pay, tax, net_pay
```

**Loop Through Employees and Generate Payroll Report**

```python
for emp in employees:
    gross, tax, net = calculate_pay(emp["hours_worked"], emp["hourly_rate"])

    print(f"Employee: {emp['name']}")
    print(f"  Hours Worked: {emp['hours_worked']}")
    print(f"  Hourly Rate: {emp['hourly_rate']}")
    print(f"  Gross Pay: Rs {gross}")
    print(f"  Tax Deducted: Rs {tax}")
    print(f"  Net Pay: Rs {net}")
    print("-" * 30)
```

**Save Payroll to a Text File**

To automate more, save payroll summaries.

```python
with open("payroll_report.txt", "w") as file:
    for emp in employees:
        gross, tax, net = calculate_pay(emp["hours_worked"], emp["hourly_rate"])
        file.write(f"Employee: {emp['name']}\n")
        file.write(f"  Hours Worked: {emp['hours_worked']}\n")
        file.write(f"  Hourly Rate: {emp['hourly_rate']}\n")
        file.write(f"  Gross Pay: Rs {gross}\n")
        file.write(f"  Tax Deducted: Rs {tax}\n")
        file.write(f"  Net Pay: Rs {net}\n")
        file.write("-" * 30 + "\n")
```
