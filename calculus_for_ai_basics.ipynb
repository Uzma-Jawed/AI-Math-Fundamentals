### üîπ What is a Function?

In mathematics, a **function** is a rule that takes an input and gives an output.

Example:
\[
f(x) = x^2
\]

```python
# in python
def f(x):
    return x**2

print(f(3))
```

___
###üîπ What is a Derivative?

A **derivative** tells us how a function changes.  
It's the **slope** of the function at any point.

For:
\[
f(x) = x^2
 -> f'(x) = 2x
\]

Let's plot it to understand better!


```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 100)
f_x = x**2
df_x = 2*x

plt.plot(x, f_x, label="f(x) = x¬≤")
plt.plot(x, df_x, label="f'(x) = 2x (Derivative)", linestyle="--")
plt.title("Function vs Derivative")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
```

üßÆ Code Cell 1: Imports + Functions

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the x range
x = np.linspace(-2, 2, 400)

# 1. f(x) = 3x^3 and its derivative f'(x) = 9x^2
f1 = 3 * x**3
f1_prime = 9 * x**2

# 2. f(x) = sin(x) and its derivative f'(x) = cos(x)
f2 = np.sin(x)
f2_prime = np.cos(x)

# 3. f(x) = e^x and its derivative f'(x) = e^x
f3 = np.exp(x)
f3_prime = np.exp(x)
```

üìä Code Cell 2: Plotting

```python
# Plotting
plt.figure(figsize=(15, 10))

# Plot for f(x) = 3x^3
plt.subplot(3, 2, 1)
plt.plot(x, f1, label='f(x) = 3x¬≥', color='blue')
plt.title('Function f(x) = 3x¬≥')
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.legend()

plt.subplot(3, 2, 2)
plt.plot(x, f1_prime, label="f'(x) = 9x¬≤", color='red')
plt.title("Derivative f'(x) = 9x¬≤")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.legend()

# Plot for f(x) = sin(x)
plt.subplot(3, 2, 3)
plt.plot(x, f2, label='f(x) = sin(x)', color='blue')
plt.title('Function f(x) = sin(x)')
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.legend()

plt.subplot(3, 2, 4)
plt.plot(x, f2_prime, label="f'(x) = cos(x)", color='red')
plt.title("Derivative f'(x) = cos(x)")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.legend()

# Plot for f(x) = e^x
plt.subplot(3, 2, 5)
plt.plot(x, f3, label='f(x) = eÀ£', color='blue')
plt.title('Function f(x) = eÀ£')
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.legend()

plt.subplot(3, 2, 6)
plt.plot(x, f3_prime, label="f'(x) = eÀ£", color='red')
plt.title("Derivative f'(x) = eÀ£")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```

üîç What it does:

*   plt.plot(x, f_x, ...): This plots the original function f(x) = x¬≤.

*   plt.plot(x, df_x, ...): This plots the derivative f‚Ä≤(x) = 2x on the same graph.

*   label=...: Gives a name for each line in the graph (used in the legend).

*   linestyle="--": Makes the derivative line dashed to visually separate it from the function line.
*   plt.title: Adds a title at the top of the graph.
*   plt.xlabel("x"): Labels the x-axis
*   plt.ylabel("y"): Labels the y-axis
*   plt.legend(): Displays a legend box that shows the labels for each plotted line:

üîµ f(x) = x¬≤

üî∂ f‚Ä≤(x) = 2x (Derivative)
*   plt.grid(True): Adds a grid in the background to make the plot easier to read.
*   plt.show():Finally displays the plot in Jupyter Notebook or Python environment.
